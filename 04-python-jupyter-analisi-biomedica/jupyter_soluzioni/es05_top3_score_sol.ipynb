```python
from pathlib import Path
import pandas as pd

# Carichiamo il dataset
csv_path = Path('..') / 'data' / 'vitali_pazienti.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
# Riutilizziamo lo score clinico
df['score_clinico'] = (
    (df['eta'] >= 70).astype(int) * 2
    + (df['bpm'] >= 100).astype(int)
    + (df['spo2'] < 95).astype(int) * 2
    + (df['sistolica'] >= 140).astype(int) * 2
    + (df['temperatura'] >= 37.5).astype(int)
)

top3 = df.sort_values('score_clinico', ascending=False).head(3)
top3[['id', 'score_clinico']]
```
