# Soluzione Esercizio 5 - Tuple ordinamento esplicito
Notebook soluzione con commenti dettagliati sul flusso.


```python
# Lista di tuple (nome, punteggio)
coppie = [('Luca', 3), ('Anna', 1), ('Marta', 2)]

# Copia della lista da ordinare
ordinate = coppie[:]
n = len(ordinate)
i = 0

# Selection sort esplicito
while i < n - 1:
    min_idx = i
    j = i + 1
    while j < n:
        if ordinate[j][1] < ordinate[min_idx][1]:
            min_idx = j
        j = j + 1
    if min_idx != i:
        tmp = ordinate[i]
        ordinate[i] = ordinate[min_idx]
        ordinate[min_idx] = tmp
    i = i + 1

print('Ordinate:', ordinate)
```
