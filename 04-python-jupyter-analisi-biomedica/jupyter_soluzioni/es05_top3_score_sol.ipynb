# Soluzione es05 top3 score


```python
from pathlib import Path
import pandas as pd

csv_path = Path('..') / 'data' / 'vitali_pazienti.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
df['score_clinico'] = (
    (df['eta'] >= 70).astype(int)*2 +
    (df['bpm'] >= 100).astype(int)*1 +
    (df['spo2'] < 95).astype(int)*2 +
    (df['sistolica'] >= 140).astype(int)*2 +
    (df['temperatura'] >= 37.5).astype(int)*1
)
df.sort_values('score_clinico', ascending=False)[['id','score_clinico']].head(3)
```
