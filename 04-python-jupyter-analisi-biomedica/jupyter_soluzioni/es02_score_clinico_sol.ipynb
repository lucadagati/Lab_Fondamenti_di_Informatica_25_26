# Soluzione Esercizio 2 - Somma media massimo
Notebook soluzione con commenti dettagliati sul flusso.


```python
# Lista di partenza
numeri = [7, 3, 12, 5, 9, 12, 1]

# Inizializziamo somma e massimo
somma = 0
massimo = numeri[0]

# Cicliamo sulla lista per aggiornare somma e massimo
for n in numeri:
    somma = somma + n
    if n > massimo:
        massimo = n

# Calcoliamo media
media = somma / len(numeri)

# Stampiamo risultati
print('Somma:', somma)
print('Media:', round(media, 2))
print('Massimo:', massimo)
```
