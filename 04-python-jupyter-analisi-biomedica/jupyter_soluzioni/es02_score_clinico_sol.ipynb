# Esercizio 2 - Score clinico
Soluzione semplificata con passaggi espliciti.


```python
from pathlib import Path
import pandas as pd

# Carichiamo il dataset
csv_path = Path('..') / 'data' / 'vitali_pazienti.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
# Calcolo score in modo leggibile (somma di blocchi)
score_eta = (df['eta'] >= 70).astype(int) * 2
score_bpm = (df['bpm'] >= 100).astype(int) * 1
score_spo2 = (df['spo2'] < 95).astype(int) * 2
score_sistolica = (df['sistolica'] >= 140).astype(int) * 2
score_temp = (df['temperatura'] >= 37.5).astype(int) * 1

df['score_clinico'] = score_eta + score_bpm + score_spo2 + score_sistolica + score_temp

# Classe rischio
def classe(score):
    if score <= 2:
        return 'Basso'
    if score <= 5:
        return 'Medio'
    return 'Alto'

df['classe_rischio'] = df['score_clinico'].apply(classe)
df[['id', 'score_clinico', 'classe_rischio']]
```
