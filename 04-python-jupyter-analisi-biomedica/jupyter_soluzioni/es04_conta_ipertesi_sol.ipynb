Notebook soluzione con commenti dettagliati sul flusso.


```python
# Funzione per contare le vocali
def conta_vocali(testo):
    count = 0
    for ch in testo.lower():
        if ch in 'aeiou':
            count = count + 1
    return count

frase = 'Programmare in Python e utile'
print('Vocali:', conta_vocali(frase))
```
