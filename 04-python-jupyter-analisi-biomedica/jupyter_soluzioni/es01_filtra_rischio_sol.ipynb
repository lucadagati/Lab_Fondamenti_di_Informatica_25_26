# Soluzione Esercizio 1 - Classifica voto


```python
def classifica_voto(voto):
    if voto < 18:
        return 'Insufficiente'
    if voto <= 24:
        return 'Sufficiente'
    if voto <= 29:
        return 'Buono'
    return 'Ottimo'

for v in [15,22,27,30]:
    print(v, '->', classifica_voto(v))
```
