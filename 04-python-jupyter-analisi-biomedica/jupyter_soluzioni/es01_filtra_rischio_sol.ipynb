# Soluzione Esercizio 1 - Classifica voto
Notebook soluzione con commenti dettagliati sul flusso.


```python
# Definiamo una funzione che classifica il voto in una categoria testuale
def classifica_voto(voto):
    # Se il voto e inferiore a 18 -> insufficiente
    if voto < 18:
        return 'Insufficiente'
    # Se il voto e compreso tra 18 e 24 -> sufficiente
    if voto <= 24:
        return 'Sufficiente'
    # Se il voto e compreso tra 25 e 29 -> buono
    if voto <= 29:
        return 'Buono'
    # Altrimenti voto ottimo
    return 'Ottimo'

# Proviamo la funzione su una lista di voti
voti = [15, 22, 27, 30]
for v in voti:
    print(v, '->', classifica_voto(v))
```
