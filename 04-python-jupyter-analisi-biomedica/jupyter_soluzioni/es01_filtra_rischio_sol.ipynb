# Esercizio 1 - Filtro pazienti a rischio
Soluzione semplificata con passaggi espliciti.


```python
from pathlib import Path
import pandas as pd

# Carichiamo il dataset
csv_path = Path('..') / 'data' / 'vitali_pazienti.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
# Regola rischio: almeno una condizione vera
df['flag_rischio'] = (df['bpm'] >= 100) | (df['spo2'] < 95) | (df['sistolica'] >= 140)

# Mostriamo solo i pazienti a rischio
df_rischio = df[df['flag_rischio']]
df_rischio[['id', 'bpm', 'spo2', 'sistolica']]
```
