# Soluzione Esercizio 8 - Min max con tuple
Notebook soluzione con commenti dettagliati sul flusso.


```python
# Funzione min_max
def min_max(valori):
    minimo = valori[0]
    massimo = valori[0]
    for v in valori:
        if v < minimo:
            minimo = v
        if v > massimo:
            massimo = v
    return minimo, massimo

nums = [4, 19, 2, 8, 11]
mn, mx = min_max(nums)
print('Minimo:', mn)
print('Massimo:', mx)
```
