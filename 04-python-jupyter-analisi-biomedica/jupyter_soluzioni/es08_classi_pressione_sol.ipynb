# Soluzione es08 classi pressione


```python
from pathlib import Path
import pandas as pd

csv_path = Path('..') / 'data' / 'vitali_pazienti.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
def classe_pressione(s):
    if s < 130:
        return 'Normale'
    if s < 140:
        return 'Pre-ipertensione'
    return 'Ipertensione'

df['classe_pressione'] = df['sistolica'].apply(classe_pressione)
df[['id','sistolica','classe_pressione']]
```
