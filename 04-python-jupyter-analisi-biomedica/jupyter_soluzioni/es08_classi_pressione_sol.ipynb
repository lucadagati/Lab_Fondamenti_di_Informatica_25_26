# Soluzione Esercizio 8 - Funzione con tuple


```python
def min_max(valori):
    mn=valori[0]
    mx=valori[0]
    for v in valori:
        if v < mn:
            mn=v
        if v > mx:
            mx=v
    return mn,mx

nums=[4,19,2,8,11]
mn,mx=min_max(nums)
print('Minimo:', mn)
print('Massimo:', mx)
```
