# Soluzione Esercizio 6 - Dizionario frequenze
Notebook soluzione con commenti dettagliati sul flusso.


```python
# Testo di input
testo = 'banana split'

# Dizionario frequenze
freq = {}

# Cicliamo su ogni carattere
for ch in testo.lower():
    if ch == ' ':
        continue
    if ch not in freq:
        freq[ch] = 0
    freq[ch] = freq[ch] + 1

print(freq)
```
