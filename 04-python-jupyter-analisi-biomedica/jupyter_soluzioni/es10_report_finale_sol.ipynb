```python
from pathlib import Path
import pandas as pd

# Carichiamo il dataset
csv_path = Path('..') / 'data' / 'vitali_pazienti.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
# Costruiamo score
df['score_clinico'] = (
    (df['eta'] >= 70).astype(int) * 2
    + (df['bpm'] >= 100).astype(int)
    + (df['spo2'] < 95).astype(int) * 2
    + (df['sistolica'] >= 140).astype(int) * 2
    + (df['temperatura'] >= 37.5).astype(int)
)

totale = len(df)
rischio_alto = int((df['score_clinico'] >= 6).sum())
idx_top = df['score_clinico'].idxmax()
top_id = df.loc[idx_top, 'id']
top_score = int(df.loc[idx_top, 'score_clinico'])

print('Totale pazienti:', totale)
print('Rischio alto (score >= 6):', rischio_alto)
print('ID score massimo:', top_id, 'score', top_score)
```
