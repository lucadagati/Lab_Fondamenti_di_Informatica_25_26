# Soluzione Esercizio 10 - Analisi parole
Notebook soluzione con commenti dettagliati sul flusso.


```python
# Frase di esempio
frase = 'python python laboratorio codice laboratorio python'

# Dizionario frequenze
freq = {}

# Conteggio parole
for parola in frase.lower().split():
    if parola not in freq:
        freq[parola] = 0
    freq[parola] = freq[parola] + 1

# Totale parole
tot = 0
for k in freq:
    tot = tot + freq[k]

# Parola con frequenza massima
parola_top = ''
freq_top = -1
for k in freq:
    if freq[k] > freq_top:
        freq_top = freq[k]
        parola_top = k

print('Frequenze:', freq)
print('Totale parole:', tot)
print('Parola top:', parola_top, '->', freq_top)
```
