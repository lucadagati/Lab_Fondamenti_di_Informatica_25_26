# Soluzione Esercizio 3 - While validazione
Notebook soluzione con commenti dettagliati sul flusso.


```python
# Simuliamo un input iniziale non valido
n = -2

# Validazione con while: ripeti finche n non diventa positivo
while n <= 0:
    # In un programma reale qui useremmo input()
    n = 5

# Stampa da 1 a n con while
i = 1
while i <= n:
    print(i)
    i = i + 1

print('Valore finale di n:', n)
```
