# Soluzione Esercizio 7 - Filtri su lista con ciclo for


```python
valori = [12, 5, 18, 21, 4, 30, 9]

# Filtro esplicito con for
filtrati = []
for v in valori:
    if v >= 10:
        filtrati.append(v)

# Quadrati con secondo ciclo esplicito
quadrati = []
for v in filtrati:
    quadrati.append(v * v)

print('Filtrati:', filtrati)
print('Quadrati:', quadrati)
```
