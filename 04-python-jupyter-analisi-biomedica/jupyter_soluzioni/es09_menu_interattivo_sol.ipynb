# Soluzione Esercizio 9 - Menu while


```python
for scelta in ['1','2','0']:
    if scelta == '1':
        n=4
        print('Quadrato:', n*n)
    elif scelta == '2':
        n=3
        print('Cubo:', n*n*n)
    elif scelta == '0':
        print('Uscita')
        break
    else:
        print('Scelta non valida')
```
