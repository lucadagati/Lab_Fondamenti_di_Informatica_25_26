```python
from pathlib import Path
import pandas as pd

# Carichiamo il dataset
csv_path = Path('..') / 'data' / 'vitali_pazienti.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
# In notebook mostriamo una simulazione delle 2 opzioni
conteggio_rischio = int(((df['bpm'] >= 100) | (df['spo2'] < 95) | (df['sistolica'] >= 140)).sum())
media_bpm = float(df['bpm'].mean())

print('Opzione 1 -> Conteggio rischio:', conteggio_rischio)
print('Opzione 2 -> Media bpm:', round(media_bpm, 2))
print('Opzione 0 -> Uscita')
```
