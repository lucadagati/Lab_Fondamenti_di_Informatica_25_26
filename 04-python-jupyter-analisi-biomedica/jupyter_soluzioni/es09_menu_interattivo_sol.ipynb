# Soluzione es09 menu interattivo


```python
from pathlib import Path
import pandas as pd

csv_path = Path('..') / 'data' / 'vitali_pazienti.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
def conteggio_rischio(frame):
    return int(((frame['bpm'] >= 100) | (frame['spo2'] < 95) | (frame['sistolica'] >= 140)).sum())

def media_bpm(frame):
    return float(frame['bpm'].mean())

# Simulazione di una scelta di menu in notebook (senza loop infinito interattivo)
print('Scelta 1 -> Conteggio rischio:', conteggio_rischio(df))
print('Scelta 2 -> Media bpm:', round(media_bpm(df), 2))
```
